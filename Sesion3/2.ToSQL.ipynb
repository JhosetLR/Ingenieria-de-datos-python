```python
def validate_duplicated(df, column_name):
    if df[column_name].duplicated().any():
        print("Hay departamentos duplicados en el dataframe.")
        exit(1)
    else:
        print("No hay departamentos duplicados en el dataframe.")
```

```python
def validate_department(df):
    if df['department_name'].duplicated().any():
        print("Hay departamentos duplicados en el dataframe.")
        exit(1)
    else:
        print("No hay departamentos duplicados en el dataframe.")
```

```python
def validate_ids(df,valids_ids):
    if not df['category_department_id'].isin(valids_ids).all():
        print ('Hay ids que no se encuentran en el dataframe')
    else:
        print('OK')
```

```python
validate_department(df_departments)
```
